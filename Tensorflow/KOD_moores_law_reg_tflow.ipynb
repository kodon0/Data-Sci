```python
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import wget
import pandas as pd
wget.download('https://raw.githubusercontent.com/lazyprogrammer/machine_learning_examples/master/tf2.0/moore.csv')
```

```python
data = pd.read_csv('moore.csv').values
```

```python
X = data[:,0].reshape(-1,1)
y = data[:,1]
```

```python
plt.scatter(X,y)
```

```python
y = np.log(y)
```

```python
# Make linear

plt.scatter(X,y)
```

```python
# Scaling X

X = X - X.mean()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Define model function and scheduler

def model_maker():
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(units = 1, input_shape = X_train.shape, activation = 'relu'))
    model.compile(optimizer = tf.keras.optimizers.SGD(learning_rate = 0.01, momentum = 0.9),loss = 'mse')
    
    return model

def schedule(epoch,lr):
    if epoch >= 50:
        return 0.0001
    return 0.001

scheduler = tf.keras.callbacks.LearningRateScheduler(schedule)
```

```python
model = model_maker()
```

```python
model.summary()
```

```python
regressor = model.fit(X_train, y_train, epochs = 500, validation_data = (X_test, y_test), callbacks=[scheduler])
```

```python
# Evaluation

print("The training loss is: ", model.evaluate(X_train,y_train))
print("The test loss is: ", model.evaluate(X_test,y_test))

plt.plot(regressor.history['loss'], label = 'loss')
plt.plot(regressor.history['val_loss'], label = 'validation loss')
plt.legend()
```

```python
# Getting slope/trained weights

print(model.layers[0].get_weights())
```
